```python
arr = np.array([[1, 0, 1, 0], [1, 0, 2, 1], [1, 0, 3, 2]])
```

```python
arr % 3 == 0
```

```python
shape = 2, 3, 4
N = np.product(shape)
arr = np.arange(N).reshape(shape)
arr
```

```python
np.nonzero(arr % 3 == 0)
```

```python
arr[np.nonzero(arr % 3 ==0)]
```

```python
arr[np.nonzero(arr % 3 ==0)] = -1
```

```python
arr
```

```python
arr == -1
```
