```python
# dropping vals with 50%+ of missing values
nan_prct_mask = df.isnull().sum() > df.shape[0] / 2
col_to_drop = nan_prct_mask[nan_prct_mask].keys()
df = df.drop(col_to_drop, axis='columns')
```

```python
# checking for discrepetnies between datasets and other missing values, filling nan values with most typical value (dominant)
cols_to_fill = ['MSZoning', 'Utilities', 'Exterior1st', 'Exterior2nd',
'BsmtFinSF1', 'BsmtFinType2', 'BsmtFinSF2', 'BsmtUnfSF',
'TotalBsmtSF', 'Electrical', 'BsmtFullBath', 'BsmtHalfBath',
'KitchenQual', 'Functional', 'GarageCars', 'GarageArea',
'SaleType']
nan_prct_mask = df.isnull().sum() > 1
col_to_drop = list(nan_prct_mask[nan_prct_mask].keys())
cols_to_fill = cols_to_fill + col_to_drop
cols_to_fill
for col in cols_to_fill:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
# no more missign values
nan_prct_mask = df.isnull().sum() > 1
col_to_drop = nan_prct_mask[nan_prct_mask].keys()
col_to_drop
```
