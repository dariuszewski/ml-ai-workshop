```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_rows', 90)
```

```python
df = pd.read_csv('house-prices-advanced-regression-techniques\\train.csv', index_col='Id')
```

```python
df
```

```python
df.isnull().sum()
```

```python
# dropping vals with 50%+ of missing values
nan_prct_mask = df.isnull().sum() > df.shape[0] / 2
col_to_drop = nan_prct_mask[nan_prct_mask].keys()
df = df.drop(col_to_drop, axis='columns')
```

```python
# checking for discrepetnies between datasets and other missing values, filling nan values with most typical value (dominant)
cols_to_fill = ['MSZoning', 'Utilities', 'Exterior1st', 'Exterior2nd',
'BsmtFinSF1', 'BsmtFinType2', 'BsmtFinSF2', 'BsmtUnfSF',
'TotalBsmtSF', 'Electrical', 'BsmtFullBath', 'BsmtHalfBath',
'KitchenQual', 'Functional', 'GarageCars', 'GarageArea',
'SaleType']
nan_prct_mask = df.isnull().sum() > 1
col_to_drop = list(nan_prct_mask[nan_prct_mask].keys())
cols_to_fill = cols_to_fill + col_to_drop
cols_to_fill
for col in cols_to_fill:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
# no more missign values
nan_prct_mask = df.isnull().sum() > 1
col_to_drop = nan_prct_mask[nan_prct_mask].keys()
col_to_drop
```

```python
# handling categorical data
train_df = df.copy()
test_df = pd.read_csv('test_df_no_nan.csv', index_col='Id')
main_df = pd.concat([train_df, test_df])
main_df
```

```python
# set categorical variables
final_df = pd.get_dummies(main_df)
final_df
```

```python
# slice back to train and test sets
df_train = final_df.iloc[:1460, :]
df_test = final_df.iloc[1460:, :]
df_test.drop(['SalePrice'], axis=1, inplace=True)
```

```python
# Building a model
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression

# X_train = df_train.drop(['SalePrice'], axis=1)
# y_train = df_train['SalePrice']

X = df_train.drop(['SalePrice'], axis=1)
y = df_train['SalePrice']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42) 
lr = LinearRegression().fit(X_train, y_train)
```

```python
print("lr.coef_: {}".format(lr.coef_)) 
print("lr.intercept_: {}".format(lr.intercept_)) 
```

```python
print("Wynik dla zestawu uczącego: {:.2f}".format(lr.score(X_train, y_train))) 
print("Wynik dla zestawu testowego: {:.2f}".format(lr.score(X_test, y_test))) 
```

```python
prediction = lr.predict(df_test)
pred=pd.DataFrame(prediction)
sub_df = pd.read_csv('house-prices-advanced-regression-techniques\\sample_submission.csv')
dataset = pd.concat([sub_df['Id'], pred], axis='columns')
dataset.columns = ['Id', 'SalePrice']
dataset.to_csv('house-prices-advanced-regression-techniques\\sample_submission.csv', index=False)
```
